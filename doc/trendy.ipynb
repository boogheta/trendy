```python
def get_a_dic(res_average):
    dico={}
    for word_stat in res_average['averages']:    
        dico[word_stat['term']]=word_stat['value']
    return dico
```

```python
def find_comparison_level(word,animal):
    res=get_temporal_profile_averages([animal,word])
    dico_data=get_a_dic(res)
    print dico_data
    clause=False
    if all(x > 0 for x in dico_data.values()):
        clause=True
    return clause, dico_data[word]/dico_data[animal]
    
```
